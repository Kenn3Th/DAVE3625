```python
import pandas as pd
import numpy as np
```

```python
df_wine = pd.read_csv('winemag-data_first150k.csv')
```

```python
df_wine.head(5)
```

```python
df_wine.plot.scatter(x='points', y='price', title='Wine')
```

```python
df_wine.drop(df_wine.columns[0], axis=1).plot.line(title='Wine Dataset')
```

```python
'''
.drop() tar vekk noe fra dataframe, 
.drop(df_wine.columns[0], axis=1) tar ut første kolonne
Separerer price og points
'''
df_points = df_wine.drop(['price', df_wine.columns[0]], axis=1)
df_price = df_wine.drop(['points', df_wine.columns[0]], axis=1)
df_points.plot.line(title='Wine points')
```

```python
df_price.plot.line(title='Wine price')
```
